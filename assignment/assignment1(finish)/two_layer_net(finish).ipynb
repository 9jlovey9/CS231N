## Answer:

 `1 Sigmoid`, `2 Relu`

$$
\text{Sigmod: } f(x)=\frac{1}{1+e^{-x}}
$$
$$
\text{Relu: } f(x)=\max(0,x)
$$

Explanation:

![Sigmoid](pic/Sigmoid.png)

* _Sigmoid_
  * 通过上图可以看出，当输入值很小或者很大的时候，sigmod的梯度都会趋向于0，而我们经常使用梯度乘以学习率来更新权值函数的，所以会导致权值改变非常小，难以训练。这种情况，可能发生在：如果数据未归一化或权重值绝对值太大。（网络层数多也会）
  * 通常我们会初始化参数的值在 $(0,1)$ 之间，通过sigmod的导数图像来看，当输入在 $(0,1)$ 之间的时候，梯度值大约在 $(0.2-0.25)$ 之间。由反向传播算法的数学推导可知，梯度从后向前传播时，每传递一层梯度值都会减小为原来的0.25倍，如果神经网络隐层特别多，那么梯度在穿过多层后将变得非常小接近于0。
  * 当网络权值初始化为 $(1,+\infty)$ 区间内的值，则会出现梯度爆炸情况
* _ReLU_ 
  * 可能会使得停止更新。若所有的输入的输出值为0，则梯度为0，反向传播失效，参数无法更小。原因可能为过大的学习率参数，使得权重系数W过小了。

Examples in 1D:
* _Sigmoid_: 当输入很大或者很小时，例如 $x = 10^5$ 或者 $x = 10^{-5}$，根据公式 $\sigma(x)=\frac{1}{1+e^{-x}}$，输出值接近 `0` or `1`。根据导数公式， $\nabla_x\sigma=(1-\sigma(x))\sigma(x)$，梯度接近 `0`. 
* _ReLU_: 根据公式 $\text{ReLU}(x)=\max(0,x)$， 如果 $x<0$，输出为 `0`，则梯度也为 `0`.
